```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('houseprice.csv')
df.head()
```

```python
df.shape
```

```python
x=df.drop(['Id','SalePrice'],axis=1)
y=df[['SalePrice']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1212)
```

```python
dfTrainNum=x_train.select_dtypes(np.number)
dfTrainCat=x_train.select_dtypes(object)
```

```python
temp=sImputer.transform(dfTrainNum)
dfTrainNaRemoved=pd.DataFrame(temp,columns=dfTrainNum.columns)
```

```python
dfTrainCatNaRemoved.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfTrainCatEncoded=dfTrainCatNaRemoved.apply(le.fit_transform)
```

```python
ss.fit(dfTrainNaRemoved)
```

```python
dfTrainNumScaled=pd.DataFrame(ss.transform(dfTrainNaRemoved),columns=dfTrainNaRemoved.columns)
```

```python
reducedDF=pd.DataFrame(pca.transform(dfTrainFinal))
```

# Cross Validation on Regression Problems

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
#X = X.to_numpy()
#y = y.to_numpy()
pca = PCA(42)
rmse = []
r2 = []
sImputerCat=SimpleImputer(strategy='most_frequent')
sImputer=SimpleImputer(strategy='mean')
le = LabelEncoder()
ss = StandardScaler()
for train_index, test_index in kf.split(x, y):
    
    # TRAIN DATA ANALYSIS
    dfTrainNum=x.iloc[train_index].select_dtypes(np.number)
    dfTrainCat=x.iloc[train_index].select_dtypes(object)
    
    ## Null Values are imputed
    sImputerCat.fit(dfTrainCat)
    dfTrainCat = pd.DataFrame(sImputerCat.transform(dfTrainCat))
    sImputer.fit(dfTrainNum)
    dfTrainNum = sImputer.transform(dfTrainNum)
    
    # Scaling and Encoding
    dfTrainCat = dfTrainCat.apply(le.fit_transform)
    ss.fit(dfTrainNum)
    dfTrainNum = pd.DataFrame(ss.transform(dfTrainNum))
    
    # Concatenate
    dfTrainFinal = pd.concat([dfTrainCat, dfTrainNum], axis=1)
    pca.fit(dfTrainFinal)
    X_train = pca.transform(dfTrainFinal)
    
    # TEST DATA ANALYSIS
    dfTestNum=x.iloc[test_index].select_dtypes(np.number)
    dfTestCat=x.iloc[test_index].select_dtypes(object)
    
    ## Null Values are imputed
    dfTestCat = pd.DataFrame(sImputerCat.transform(dfTestCat))
    dfTestNum = sImputer.transform(dfTestNum)
    
    # Scaling and Encoding
    dfTestCat = dfTestCat.apply(le.fit_transform)
    dfTestNum = pd.DataFrame(ss.transform(dfTestNum))
    
    # Concatenate
    dfTestFinal = pd.concat([dfTestCat, dfTestNum], axis=1)
    X_test = pca.transform(dfTestFinal)
    
    training = rf.fit(X_train, y.iloc[train_index])
    
    pred = rf.predict(pca.transform(dfTestFinal))
    
    rmse.append(
        np.sqrt(mean_squared_error(pred, y.iloc[test_index])
    ))
    r2.append(r2_score(pred, y.iloc[test_index]))
```
